# Querying Available Data Sets

1,300 data sets in all via their API. Some, like 
the *American Community Survey: 5-Year Estimates: Data Tables*
are widely known and used, but there are plenty of other interesting and useful
data sets you might want to load.

`censusdis` supports querying avaialble data sets for a given year or for all time.

We have not loaded all variables for all data sets for all years, but

## How many datasets by year.

Some older years do not have very many vintage data sets. Note also that we have some `'timeseries'` datasets that are multi-year.

We will plot which of the most popular datasets are available in which years, and how many other datasets there are.

```python
datasets_to_count = {
    "ACS 1 Year": "acs/acs1",
    "ACS 3 Year": "acs/acs3",
    "ACS 5 Year": "acs/acs5",
    "Population Estimates": "pep/population",
    "Decennial Redistricting": "dec/pl",
}

s_counts_by_dataset = {
    name: df_all_datasets[df_all_datasets.DATASET == dataset]
    .groupby("YEAR")["DATASET"]
    .count()
    for name, dataset in datasets_to_count.items()
}



s_counts_other = {
    "Other": df_all_datasets[~df_all_datasets.DATASET.isin(datasets_to_count)]
    .groupby("YEAR")["DATASET"]
    .count()
}

# Make all the indices strings so when pandas sorts them it does not
# have a problem comparing ints and strings.
s_counts_other["Other"].index = s_counts_other["Other"].index.map(str) 
for s_counts in s_counts_by_dataset.values():
    s_counts.index = s_counts.index.map(str)
    
    
df_years = pd.DataFrame([str(year) for year in range(1986, 2022)])

df_dataset_counts = df_years.join(
    pd.DataFrame(dict(s_counts_by_dataset, **s_counts_other)), how="outer"
)

df_dataset_counts
```

```python
s_counts_other['Other'].index
```

```python
s_counts_by_dataset['ACS 1 Year'].index
```

```python
ax = df_dataset_counts[datasets_to_count.keys()].plot.bar(stacked=True, figsize=(12, 8))

ax.set_title("Selected Popular Datasets by Vintage")

ax.set_yticks(range(5))

ax.grid(axis="y")

None
```

```python
ax = df_dataset_counts.plot.bar(stacked=True, figsize=(12, 8))

ax.set_title("All Available Datasets by Vintage")

ax.grid(axis="y")
```

## List the groups in a dataset

```python
dataset = df_datasets_for_year.iloc[4]["DATASET"]
dataset
```

```python
df_groups = ced.variables.all_groups(dataset, YEAR)

df_groups
```

## List the variables in a group

```python
group = df_groups.iloc[-5]["GROUP"]
group
```

```python
df_variables = ced.variables.all_variables(dataset, YEAR, group)
df_variables
```

## Show the variables in a tree

```python
ced.variables.group_tree(dataset, YEAR, group)
```

## List all the leaves of the tree

```python
leaves = ced.variables.group_leaves(dataset, YEAR, group)
leaves
```

## Query the data

```python
# Note that these two are equivalent. Using `leaves_of_group=group`
# could have saved us needing to construct `leaves` with the
# separate call above.

df_download_variables = ced.download(dataset, YEAR, leaves, state="*")
df_download_leaves_of_group = ced.download(
    dataset, YEAR, leaves_of_group=group, state="*"
)

df_download_leaves_of_group.equals(df_download_variables)
```
