# MA Remove Water

This notebook demonstrates case where `remove_water=True` removes a large
number of small inland bodies of water.

```python
import censusdis.data as ced
import censusdis.maps as cem
from censusdis.datasets import ACS5
from censusdis.states import MA, CT, RI, NH, NAMES_FROM_IDS
```

```python
STATE = MA
```

Load county level data with and without removing water.

```python
gdf_county = ced.download(
    ACS5,
    2020,
    ["NAME"],
    state=STATE,
    county="*",
    with_geometry=True,
)
```

```python
gdf_county_remove_water = ced.download(
    ACS5,
    2020,
    ["NAME"],
    state=STATE,
    county="*",
    with_geometry=True,
    remove_water=True,
)
```

Plot both of them.

```python
ax = cem.plot_map(
    gdf_county,
    color="green",
    alpha=0.25,
    edgecolor="black",
    linewidth=2,
    with_background=True,
    figsize=(12, 6),
)

_ = ax.set_title(f"{NAMES_FROM_IDS[STATE]} Counties")
```

```python
ax = cem.plot_map(
    gdf_county_remove_water,
    color="purple",
    alpha=0.25,
    edgecolor="black",
    linewidth=2,
    with_background=True,
    figsize=(12, 6),
)

_ = ax.set_title(f"{NAMES_FROM_IDS[STATE]}  Counties: remove_water=True")
```
